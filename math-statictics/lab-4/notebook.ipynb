```python
# =============================
# ЗАДАНИЕ 4 — Математическая статистика
# Полный код для Jupyter Notebook
# =============================

import numpy as np
import scipy.stats as stats
import pandas as pd

np.random.seed(42)

def print_decision(p_value, alpha=0.05):
    if p_value < alpha:
        print("Отвергаем H0")
    else:
        print("Нет оснований отвергнуть H0")

alpha = 0.05  # уровень значимости (1 - 0.95)

# =============================
# 1.1.1 Некоррелированность
# =============================
X = np.random.normal(6, np.sqrt(3), 500)
X1, X2 = np.split(X, 2)

corr, p_value = stats.pearsonr(X1, X2)
print("1.1.1 Pearson corr:", corr, "p-value:", p_value)
#print_decision(p_value, alpha)

# =============================
# 1.1.2 Связанная выборка
# =============================
eps = np.random.normal(0, 1, 500)
Y = 4 * X - 9 + eps

X1, Y1 = np.split(X, 2)[0], np.split(Y, 2)[0]
corr, p_value = stats.pearsonr(X1, Y1)
print("1.1.2 Pearson corr:", corr, "p-value:", p_value)
#print_decision(p_value, alpha)

# =============================
# 1.2.1 t-критерий (независимые выборки)
# =============================
X1 = np.random.normal(8, np.sqrt(6), 500)
A, B = np.split(X1, 2)

stat, p_value = stats.ttest_ind(A, B)
print("1.2.1 t-test:", stat, "p-value:", p_value)
#print_decision(p_value, alpha)

# =============================
# 1.2.2 разные средние
# =============================
X2 = np.random.normal(4, np.sqrt(5), 500)
stat, p_value = stats.ttest_ind(X1, X2)
print("1.2.2 t-test:", stat, "p-value:", p_value)
#print_decision(p_value, alpha)

# =============================
# 1.2.3 близкие средние
# =============================
X1 = np.random.normal(8, np.sqrt(6), 500)
X2 = np.random.normal(8.3, np.sqrt(5), 500)

stat, p_value = stats.ttest_ind(X1, X2)
print("1.2.3 t-test:", stat, "p-value:", p_value)
#print_decision(p_value, alpha)

# =============================
# 1.3.1 Связанные выборки
# =============================
X = np.random.normal(5, np.sqrt(5), 500)
A, B = np.split(X, 2)

stat, p_value = stats.ttest_rel(A, B)
print("1.3.1 paired t-test:", stat, "p-value:", p_value)
#print_decision(p_value, alpha)

# =============================
# 1.3.2 Добавляем шум
# =============================
eps = np.random.normal(0, 1, 500)
Y = X + eps

A, B = np.split(X, 2)[0], np.split(Y, 2)[0]
stat, p_value = stats.ttest_rel(A, B)
print("1.3.2 paired t-test:", stat, "p-value:", p_value)
#print_decision(p_value, alpha)

# =============================
# 2.1 Критерий Фишера
# =============================
X1 = np.random.normal(2, np.sqrt(6), 500)
A, B = np.split(X1, 2)

F = np.var(A, ddof=1) / np.var(B, ddof=1)
dfn, dfd = len(A)-1, len(B)-1
p_value = 1 - stats.f.cdf(F, dfn, dfd)
print("2.1 F-test:", F, "p-value:", p_value)
#print_decision(p_value, alpha)

# =============================
# 2.2.1 разные дисперсии
# =============================
X2 = np.random.normal(1, np.sqrt(9), 750)

F = np.var(X1, ddof=1) / np.var(X2, ddof=1)
dfn, dfd = len(X1)-1, len(X2)-1
p_value = 1 - stats.f.cdf(F, dfn, dfd)
print("2.2.1 F-test:", F, "p-value:", p_value)
#print_decision(p_value, alpha)

# =============================
# 2.2.2 близкие дисперсии
# =============================
X1 = np.random.normal(1, np.sqrt(8.7), 500)
X2 = np.random.normal(1, np.sqrt(9), 750)

F = np.var(X1, ddof=1) / np.var(X2, ddof=1)
dfn, dfd = len(X1)-1, len(X2)-1
p_value = 1 - stats.f.cdf(F, dfn, dfd)
print("2.2.2 F-test:", F, "p-value:", p_value)
#print_decision(p_value, alpha)

# =============================
# 3.1 Пуассон — однородность
# =============================
X1 = np.random.poisson(2, 500)
A, B = np.split(X1, 2)

lambda1 = np.mean(A)
lambda2 = np.mean(B)

stat = (lambda1 - lambda2) / np.sqrt(lambda1/len(A) + lambda2/len(B))
p_value = 2 * (1 - stats.norm.cdf(abs(stat)))

print("3.1 Poisson test:", stat, "p-value:", p_value)
#print_decision(p_value, alpha)

# =============================
# 3.2 разные λ
# =============================
X2 = np.random.poisson(3, 500)

lambda1 = np.mean(X1)
lambda2 = np.mean(X2)

stat = (lambda1 - lambda2) / np.sqrt(lambda1/len(X1) + lambda2/len(X2))
p_value = 2 * (1 - stats.norm.cdf(abs(stat)))

print("3.2 Poisson test:", stat, "p-value:", p_value)
#print_decision(p_value, alpha)

```
